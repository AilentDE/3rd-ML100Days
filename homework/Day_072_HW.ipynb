# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def ReLU(x):
    y = []
    for i in x:
        y.append(max(0, i))
    return y

def dReLU(x):
    y = []
    for i in x:
        if i >= 0:
            y.append(1)
        else:
            y.append(0)
    return y

x = plt.linspace(-10, 10, 100)

plt.figure(figsize=(10,5))
plt.plot(x, ReLU(x), 'b', label='ReLU', alpha=0.7)
plt.plot(x, dReLU(x), 'r', label='dReLU', alpha=0.7)
plt.grid()
plt.title('ReLU & dReLU Function')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.legend(loc='best', fontsize=13, shadow=True, title='Fuction')

plt.show()
```
