```python
import pandas as pd
import numpy as np
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
data = pd.DataFrame(digits['data'])
target = pd.DataFrame(digits['target'], columns=['Target'])
data = pd.merge(data, target, left_index=True, right_index=True)
del target
data.head(10)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data['Target'], test_size=0.25, random_state=42)

RFmodel = RandomForestClassifier(n_estimators=100)
RFmodel.fit(x_train, y_train)
y_pred = RFmodel.predict(x_test)

print('The Accuracy score of RandomForestClassifier :',metrics.accuracy_score(y_test, y_pred))
```

```python
GBmodel = GradientBoostingClassifier()
GBmodel.fit(x_train, y_train)
y_pred = GBmodel.predict(x_test)

print('Te Accuracy score of GradientBoostingClassifier :',metrics.accuracy_score(y_test, y_pred))
```
