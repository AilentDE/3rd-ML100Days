- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(15,5))
plt.plot(range(1,11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Within cluster sum of squares')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
y_kmeans = kmeans.fit_predict(X)
```

```python
print('K-means:\n', y_kmeans, sep='')
print('True:\n', y, sep='')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X[y_kmeans==0, 0], X[y_kmeans==0, 1], s=100, color='r', label='Target 0', alpha=0.7)
sns.scatterplot(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=100, color='b', label='Target 1', alpha=0.7)
sns.scatterplot(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=100, color='g', label='Target 2', alpha=0.7)
#plotting centroids of the clusters
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, color='y', marker='*', label='Centroids')
plt.title('K-means', size=25)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()


plt.figure(figsize=(10,8))
sns.scatterplot(X[y==0, 0], X[y==0, 1], s=100, color='r', label='Target 0', alpha=0.7)
sns.scatterplot(X[y==1, 0], X[y==1, 1], s=100, color='b', label='Target 1', alpha=0.7)
sns.scatterplot(X[y==2, 0], X[y==2, 1], s=100, color='g', label='Target 2', alpha=0.7)
plt.title('True', size=25)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```
